# Introduction to Streamlit
- Streamlit is an open-source Python library that allows to create web applications for data science and machine larning projects quickly.It is designed for simplicity and ease of use, making it ideal for beginners.
- With Streamlit, we not need to write HTML, CSS, or JavaScript- it works entirely with Python.

## How Streamlit Works

- Streamlit runs a Python script and converts it into an interactive web application.
- It uses simle `streamlit run our_script.py` command to launch the applications.
- Every time a user interacts with the app(e.g., clicks a button, enters text), Streamlit automatically reruns the script from top to bottom.
- Streamlit provides built-in widgets like sliders, buttons, and text inputs for easy user interaction.

## Step-by-Step Guide to Using Streamlit

**Step 1 : Install Streamlit** : First, we need to install Streamlit in our Python environment.
- Run the following command in our terminal or command prompt.

```
pip install streamlit
```

**Step2 : Create a Simple Streamlit App** :  Create a new python script called `app.py` and write the following code.

```
import streamlit as st # Importing streamlit library

#title of the web app
st.title("My First streamlit App")

#Header
st.header("Welcome to Streamlit")

#Subheader
st.subheader("This is a beginner-Friendly tutorial")

#Display an integer number
st.write("The sum of 2+2 is:",2+2)

#Displaying a markdown text
st.markdown("### Streamlit is *amazing!")

#run this script using:streamlit run app.py
```

**Step3: Run our Streamlit App**

- After writing the script, save the file as `app.py` and run the following command in the terminal

```
streamlit run app.py
```

**Step4 : Adding User Input** : We can make the app interactive by adding widgets like text input and buttons.

```
import streamlit as st

#title
st.title("User input example")

#text input
name=st.text__input("enter your name")

#button
if st.button("Submit"):
    st.write("Hello,",name, "Welcome to streamlit!")
```

**Step5: Adding Widgets(Sliders, Selected Boxes, and Checkboxes)**: Streamlit provides built-in UI elements to make our app interactive.

```
import streamlit as st
st.title("Streamlit widgets example")

#Slider 
age=st.slider("Selected your age:",1,100,25)
st.write("Your selected age is:",age)

#selected box
gender=st.selectbox("Select your gender",["Male","Female","Other"])
st.write("You Selected",gender)

#checkbox
if st.checkbox("Show secret message"):
    st.write("Hello! you found the secret message")

```

**Step6: Displaying DataFrames and Charts**: Streamlit makes it easy to display table and charts.

```
import streamlit as st
import pandas as pd
import numpy as np

st.title("Data and charts in Streamlit")

#generate a random dataframe

df=pd.DataFrame("np.random.randn(10,3),columns=["A","B","C"])

#Displaying DataFrame
st.write("### Data Table")
st.dataframe(df)

#line chart
st.write("### Line chart")
st.line_chart(df)
```

**Step 7 : Uploading and Displaying Files**  : We can allow users to upload files using `st.file_uploader.`

```
import streamlit as st
import pandas as pd

st.title("File upload example")

#file uploader
uploaded_file=st.file_uploader("Upload a csv file",type=["csv"])

if uploaded_file is not None:
    df=pd.read_csv(uploaded_file)
    st.write("### uploaded Data")
    st.dataframe(df)

```
